```python
import numpy as np
import pandas as pd 
from pandas import Series,DataFrame
```

```python
score_list = np.random.randint(25,100,size = 20)
type(score_list)
```

```python
score_list
```

```python
bins = [0,59,70,80,100]
score_cat = pd.cut(score_list,bins)
print(type(score_cat))
score_cat
```

```python
pd.value_counts(score_cat)
```
