```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
%matplotlib inline
```

```python
import seaborn as sns
```

```python
s1 = Series(np.random.randn(1000))
```

```python
plt.hist(s1)  # 直方图 
```

```python
s1.plot(kind = "kde")
```

# seaborn

```python
sns.distplot(s1,hist = True,kde =True,rug = True)
```

```python
sns.distplot(s1,bins = 30, hist = True,kde =True,)
```

```python
sns.kdeplot(s1,shade = True,color = "red")
```

```python
sns.plt.plot(s1)
```
