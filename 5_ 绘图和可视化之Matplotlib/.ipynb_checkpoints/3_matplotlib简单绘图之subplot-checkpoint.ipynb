```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.plot(x,y1,"b--",label="sin(pi*x)")
plt.ylabel("y1 value")
plt.plot(x,y2,'r--',label = "sin(pi*2x)")
# plt.ylabel("y2 value")
plt.xlabel("x value")
plt.title("shis is x-y value")
plt.legend()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,'b--')
plt.ylabel("y1")
plt.xlabel("x1")
plt.subplot(2,2,2)
plt.plot(x,y2,"r--")
plt.ylabel("y2")
plt.xlabel("x2")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y2,"r--")
plt.ylabel("y1")
plt.xlabel("x1")
plt.subplot(2,2,2)
plt.plot(x,y2,"r--")
plt.ylabel("y2")
plt.xlabel("x2")
plt.subplot(2,2,3)
plt.plot(x,y2,"r--")
plt.ylabel("y3")
plt.xlabel("x3")
plt.subplot(2,2,4)
plt.plot(x,y2,"r--")
plt.ylabel("y4")
plt.xlabel("x4")
```
