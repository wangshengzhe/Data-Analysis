```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
s = Series([1,2,3,4,5])
s
```

```python
s.cumsum()  # 累加，1！2! 
```

```python
s1 = Series(np.random.randn(10)).cumsum()
s2 = Series(np.random.randn(10)).cumsum()
s1.plot(kind = "bar")
```

```python
s1.plot(kind = "line",grid = True,label = "s1",title = "shis is Series",style = "--")
s2.plot(label = "s2")
plt.legend()
```

```python
fig,ax = plt.subplots(2,1)
print(fig)
ax[0].plot(s1)
ax[1].plot(s2)
plt.show()
```
