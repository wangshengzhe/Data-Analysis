```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
```

```python
df = DataFrame(
     np.random.randint(1,10,40).reshape(10,4),
     columns = ["A","B","C","D"]
)
df
```

```python
df.plot(kind = "bar")
```

```python
df.plot(kind="line")  # 默认是按列画图
```

```python
df.plot(kind = "barh")
```

```python
df.plot(kind="bar",stacked = True)
```

```python
df.plot(kind = "area")
```

```python
a = df.iloc[5]  # index[5]的一行取值 
a
```

```python
type(a)
```

```python
a.plot()  # y轴就是元素的数值
```

```python
for i in df.index:  # 通过for循环实现按行画图
    df.iloc[i].plot(label =str(i) )
plt.legend()
```

```python
df["A"].plot() # 特定的列画图
```
